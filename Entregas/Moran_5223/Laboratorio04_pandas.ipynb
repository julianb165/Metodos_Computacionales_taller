a. Leer el dataset desde el link sugerido.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/hernansalinas/Metodos_Computacionales_taller/main/Asignaciones/Laboratorios/dataset/Temperature_Bogota_Barcelona.csv")
df.head()
```

b. Eliminar las columnas Month, day y Unnamed

```python
df['City'].unique()
```

```python
dato_limpio = df.drop(columns=['Unnamed: 0', 'Day', 'Month'])
dato_limpio.head()
```

Adicionar una nueva columna AvgTempCelcius con la temperatura en grados celcius. 
$C=(F-32)/1.8$

```python
dato_limpio['AvgTempCelcius'] = (dato_limpio['AvgTemperature'] - 32)/(1.8)
dato_limpio.head()
```

d. Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Bogotá.

e. Realizar un grafico de AvgTempCelcius como funcion del año para la ciudad de Barcelona.

```python
dato_bogota = dato_limpio[dato_limpio['City'] == 'Bogota']
dato_barcelona = dato_limpio[dato_limpio['City'] == 'Barcelona']
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,4))
ax[0].plot(dato_bogota.Year, dato_bogota.AvgTempCelcius)
ax[0].scatter(dato_bogota.Year, dato_bogota.AvgTempCelcius)
ax[1].plot(dato_barcelona.Year, dato_barcelona.AvgTempCelcius)
ax[1].scatter(dato_barcelona.Year, dato_barcelona.AvgTempCelcius)

ax[0].set_xlabel('Año')
ax[0].set_ylabel('Temperatura en ºC')
ax[0].set_title('''Bogotá
Gráfica Temperatura vs Año''')
ax[0].grid()
ax[1].set_xlabel('Año')
ax[1].set_ylabel('Temperatura en ºC')
ax[1].set_title('''Barcelona
Gráfica Temperatura vs Año''')
ax[1].grid()
```
