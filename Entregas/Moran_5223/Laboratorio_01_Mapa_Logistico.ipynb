##Introduccion

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(x, a, b, c):
  return a*np.sin(x)+b*np.cos(x)+x**5
```

```python
#Parametrizar
xini = 0
xfin = 5
N = 100
x = np.linspace(xini, xfin, N)
params = (1,1,2)
print(x)
```

```python
f = np.zeros(len(x))
```

```python
i = 0
for i in range (0, len(x)):
  f[i] = fun(x[i], *params)
plt.plot(x, f)
```

```python
from matplotlib.pyplot import tight_layout
plt.figure(figsize=(4,4))
plt.plot(x, f, "v-", label=f"$a\\sin bx+c$")
plt.xlabel("X", fontsize=14)
plt.ylabel("Y", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title("Funcion trigonometrica")
plt.savefig("funciontrigonometrica.png", thight=True)
plt.legend(loc=1)
```

1. Programar la ecuación para el mapa logístico, y realizar una grafica del valor de xn como funcion de n


\begin{equation}
x_{n+1} = r(1-x_{n+1})
\end{equation}

```python
def fun(r, xn):
  return r*xn*(1-xn)
```

<br/>
<br/>

2 y 3. Realizar pruebas para un x?i=0.5, con valores de r y construir una rutina llamada logist_map en que entrada el valor de r, la condiciones inicial xi y el número de puntos determine la evolución en el punto N

```python
xo = 0.5
r = 0.5
Np = 10

def logistimap(xo, r, Np):
  f = np.zeros(Np)

  for i in range(0, Np):
    xn = fun(r, xo)
    f[i] = xo
    xo = xn
  return f
  
```
